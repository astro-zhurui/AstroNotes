```python
from astropy import units as u
import matplotlib.pyplot as plt
plt.style.use('science')
```

## 在图中画出物理量
* 无需转换，可以直接画出astropy的物理量
* 这样之后画图都可以使用物理量，如果只想在当前命令中使用，可以使用with语句，[参见文档](https://docs.astropy.org/en/stable/visualization/matplotlib_integration.html)
* 物理量的单位会直接在坐标轴显示（使用场景，标度图像的坐标）

```python
from astropy.visualization import quantity_support
quantity_support()  
```

```python
plt.figure(figsize=(5, 5))
x = [1, 2, 3] * u.m
y = [0.1, 0.2, 0.3] * u.s
plt.plot(x, y)
```

## 在图中画出时间

```python
from astropy.visualization import time_support
from astropy.time import Time
time_support() 
```

```python
plt.figure(figsize=(5,3))
plt.plot(Time([58000, 59000, 62000], format='mjd'), [1.2, 3.3, 2.3])
```

```python
time_support(format='mjd', scale='tai')
plt.figure(figsize=(5,3))
plt.plot(Time([50000, 52000, 54000], format='mjd'), [1.2, 3.3, 2.3])
```
