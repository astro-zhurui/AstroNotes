```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True  #设置LaTeX字体为默认字体
```

# 插入水平线和垂直线、水平填充与垂直填充

## P1 插入水平线和垂直线
### 水平线 plt.hlines(0, xmin=0, xmax=10, linestyles='--', color='royalblue', label='hlines')
### 垂直线 plt.vlines(2, ymin=-2, ymax=2, linestyles='-.', color='orange', label='vlines')

```python
N = 25

x = np.linspace(0, 10, N)
y = np.sin(x)**2 + np.cos(x)

np.random.seed(123)
noise = np.random.random(N) * .7 + .4

plt.figure(figsize=(10, 7))
plt.plot(x, y, linestyle='-', label = r'$y = sin^2(x) + cos(x)$', color = 'darkgreen')
plt.fill_between(x, y+noise, y-noise, color = 'darkgreen', alpha= .5)
plt.axis('equal')

plt.hlines(0, xmin=0, xmax=10, linestyles='--', color='royalblue', label='hlines')
# 绘制水平线y=0, x范围(0, 10)

plt.vlines(2, ymin=-2, ymax=2, linestyles='-.', color='orange', label='vlines')
# 绘制垂直线x=2, y范围(-2, 2)

plt.xlabel('$x$(rad)')
plt.legend()
```

## P2 垂直填充与水平填充
###
###

```python
N = 25

x = np.linspace(0, 10, N)
y = np.sin(x)**2 + np.cos(x)

np.random.seed(123)
noise = np.random.random(N) * .7 + .4

plt.figure(figsize=(10, 7))
plt.plot(x, y, linestyle='-', label = r'$y = sin^2(x) + cos(x)$', color = 'darkgreen')
plt.fill_between(x, y+noise, y-noise, color = 'darkgreen', alpha= .5)
# 这句是填充误差区间
plt.axis('equal')

plt.fill_between((2,4), -10, 3.2, facecolor='orange', alpha=0.4)
# 这句才是垂直填充

plt.fill_betweenx((-2,-1), 1, 8, facecolor='blue', alpha=0.3)
# 水平填充

plt.xlim(0, 10)
plt.ylim(-3, 3)

plt.xlabel('$x$(rad)')
plt.legend()
```
