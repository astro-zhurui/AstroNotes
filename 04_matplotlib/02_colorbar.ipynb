```python
import numpy as np
import matplotlib.pyplot as plt
```

实验数据

```python
def data():
    data_x = np.linspace(-10, 10, 200)
    data_y = np.linspace(-10, 10, 200)
    x, y = np.meshgrid(data_x, data_y)
    z = np.sqrt(x**2 + y**2)
    return x, y, z
```

```python
x, y, z = data()

# 划定子图区域
fig = plt.figure(figsize=(10, 6))
ax1 = fig.add_axes([0.1, 0.2, 0.4, 0.8])
ax2 = fig.add_axes([0.5, 0.2, 0.4, 0.8])
ax_cbar = fig.add_axes([0.1, 0.1, 0.8, 0.05])

contour1 = ax1.contourf(x, y, z)
contour2 = ax2.contourf(x, y, z)
fig.colorbar(contour1,cax=ax_cbar,orientation='horizontal')
```
