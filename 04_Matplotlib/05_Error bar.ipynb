```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True  #设置LaTeX字体为默认字体
```

# Error bar

## P1 折线图中添加error bar
### plt.errorbar(x, y, xerr=noise_x, yerr=noise_y)

```python
# 生成数据点
N = 25
x = np.linspace(0, 10, N)
y = np.sin(x)**2 + np.cos(x)

np.random.seed(1)
noise_x = np.random.random(N) * .2 + .1
noise_y = np.random.random(N) * .7 + .4

plt.figure(figsize=(10, 6))
plt.errorbar(x, y, xerr=noise_x, yerr=noise_y, label=r'$y=sin^2(x)+cos(x)$',fmt='o', color='b', ecolor='g')
# fmt='o'  shape of the data point
# color='r'  color of the data point
# ecolor='k'  color of the error bar
plt.axis('equal')
plt.legend()
plt.xlabel('$x$(rad)')
```

## P2 填充误差区域
### plt.fill_between(x, y+noise, y-noise, alpha=0.2)

```python
N = 25
x = np.linspace(0, 10, N)
y = np.sin(x)**2 + np.cos(x)

np.random.seed(2)
noise = np.random.random(N) * .7 + .4

plt.figure(figsize=(10, 6))
plt.plot(x, y, ls='-.', label=r'$y=sin^2(x)+cos(x)$')
plt.fill_between(x, y+noise, y-noise, alpha=0.2)
# alpha参数：调节填充区域的不透明度

plt.axis('equal')
plt.legend()
plt.xlabel('$x$(rad)')


```
