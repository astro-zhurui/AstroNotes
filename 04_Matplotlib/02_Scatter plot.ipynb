```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True  #设置LaTeX字体为默认字体
```

# Scatter plot

## P1 基本操作
### 生成x和y：x是0到10的50个数，y是x的一个函数 绘制散点图

```python
N = 50
x = np.linspace(0,10,N)
y = np.sin(x)**2 + np.cos(x)

plt.figure(dpi = 150)
plt.scatter(x,y)
```

## P2 基本流程
### 保存图片操作

```python
plt.figure(dpi = 150)
plt.scatter(x, y, s=15, label = r'$y = sin^2(x) + cos(x)$', color = 'g')
#s表示点的大小,其中r的作用是开启LATEX模式
plt.legend()  #显示图例（label）
plt.axis('equal')  #使坐标轴比例为1:1
plt.xlabel(r'$x(rad)$')  #设置x轴标签
plt.ylabel(r'$y$')  #设置y轴标签
#plt.savefig('scatter_P2.png', dpi = 1000, bbox_inches = 'tight', facecolor = 'w')  #保存图片
```

## P3 双散点图 

```python
plt.figure(dpi = 150)
plt.subplot(121)  #创建1行2列的子图中的第1个
plt.scatter(x, y, s = 15, label = r'$y = sin^2(x) + cos(x)$', color = 'r', marker = 'x')
plt.subplot(122)  #创建1行2列的子图中的第1个
plt.scatter(x, y, s = 20, label = r'$y = sin^2(x) + cos(x)$', color = 'b', marker = r'$\rho$')
plt.show()
```

## P4 自定义每个数据点大小

```python
np.random.seed(1)

N = 5

plt.figure(figsize = (14, 6))

#################################################################
plt.subplot(1, 2, 1)  #sub图1
x = np.random.random(N) * 100
y = np.random.random(N) * 100

size = np.random.randint(100, 500, size=N)  #从50-200之间生成N个随机整数，对应刚才生成的50个数据点的大小

plt.scatter(x, y, s=size, color='darkorange')
plt.scatter(10, 10, s=800, color='orange') #将（10， 10）点在图中

plt.axis('equal')
plt.xlabel('x')
plt.ylabel('y')

#################################################################
plt.subplot(1, 2, 2)  #sub图2
x = np.random.random(N) * 100
y = np.random.random(N) * 100

size = np.random.randint(100, 500, size=N)  #从50-200之间生成N个随机整数，对应刚才生成的50个数据点的大小

plt.scatter(x, y, s=size, color='b')
plt.scatter(10, 10, s=800, color='r') #将（10， 10）点在图中

plt.axis('equal')
plt.xlabel('x')
plt.ylabel('y')

from matplotlib.ticker import MultipleLocator  #设置次刻度
ax = plt.gca()
ax.xaxis.set_minor_locator(MultipleLocator(5))
ax.yaxis.set_minor_locator(MultipleLocator(10))


#################################################################


```

## P5 带colorbar的散点图
- 普通散点图只有x，y两个坐标信息；P4中给每个点增加了大小size信息；P5则在P4的基础上增加了颜色信息

```python
np.random.seed(2)  # 不撒随机数种子，则每次运行的结果都会变化

N = 20

x = np.random.random(N) * 100
y = np.random.random(N) * 100
size = np.random.randint(20, 500, size = N)  # size同时表征了点的大小与颜色

plt.figure(figsize=(7, 5))
plt.scatter(x, y, s = size, c = size)  # s与c都等于size，即用颜色来刻画大小，点越大越黄，点越小越紫
plt.axis('equal')

ax = plt.gca()
ax.xaxis.set_minor_locator(MultipleLocator(10))
ax.yaxis.set_minor_locator(MultipleLocator(10))

plt.xlabel("x")
plt.ylabel("y")

plt.colorbar()
size
```

## P6 更改colorbar的色带
### P5中的colorbar的色带是从紫色到黄色，我们并没有设置；因此P6将介绍自定义colorbar色带颜色的方法

```python
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap

top = cm.get_cmap('Oranges_r', 128)
bottom = cm.get_cmap('Blues', 128)

newcolors = np.vstack((top(np.linspace(0, 1, 128)), bottom(np.linspace(0, 1, 128))))
orange_blue = ListedColormap(newcolors, name='OrangeBlue')

########## 以上完成了新颜色的建立 ##########

np.random.seed(1)  # 前期准备，数据生成

N = 300

x = np.random.random(N) * 100
y = np.random.random(N) * 100
size = np.random.randint(20, 800, size=N)

plt.figure(figsize=(15, 10))  # 建立图窗，开始画图
plt.scatter(x, y, s=size, c=size, alpha= .8, cmap=orange_blue)  #s表示点的大小，c表示点的颜色值
plt.axis('equal')

ax = plt.gca()  # 设置坐标轴子刻度线
ax.xaxis.set_minor_locator(MultipleLocator(10))
ax.yaxis.set_minor_locator(MultipleLocator(10))

plt.xlabel(r'$x$', fontsize=30)  # 设置坐标轴标签,设置坐标轴label字体大小
plt.ylabel(r'$y$', fontsize=30)
plt.tick_params(labelsize='20') # 设置刻度线数字字体大小

plt.colorbar(label='color size')  # 设置colorbar的标签，现在还知道怎样调整colorbar的label字体大小

plt.show()


```
