```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True  #设置LaTeX字体为默认字体
```

# Line plot

## P1 基本操作

```python
N = 100
x = np.linspace(0., 10., N)  # 在[0,10]之间均匀取N个数
y = np.sin(x) ** 2 + np.cos(x)

plt.figure(figsize=(15, 7))
plt.subplot(1,2,1)
plt.plot(x, y)  # 线图就是将离散的数据点连起来
plt.subplot(1,2,2)
plt.scatter(x, y)  # 与散点图对比
```

## P2 自定义线条样式
### plt.plot(x, y, linestyle='--')

```python
N = 50

x = np.linspace(0, 10, N)
y = np.sin(x) ** 3 + np.cos(x)

plt.figure()
plt.plot(x, y, linestyle='--')  #在plt.plot()中添加linestyle参数
```

```python
########## 自动生成不同线条的子图 ##########

#step1.划分子图的网格
rows = 2
columns = 2
grid = plt.GridSpec(rows, columns, wspace=0.55, hspace=0.5)  
# 将figure划分成有rows行，columns列的网格grid；wspace为子图宽度间隔，hspace为高度间隔
print("grid[0]的位置：",grid[0], "#用2*2的矩阵来描述子图的位置" \
    "\ngrid[1]的位置：",grid[1], \
        "\ngrid[2]的位置：",grid[2],\
            "\ngrid[3]的位置：",grid[3])

#step2.绘图
plt.figure(figsize=(9, 9))
linestyles = ['-', '--', '-.', ':']

for i in range(len(linestyles)):
    plt.subplot(grid[i])  # 指定第i个子图的位置
    plt.plot(x, y, linestyle=linestyles[i], label=r'$sin^3(x)+cos(x)$')
    plt.axis('equal')
    plt.xlabel(r'$x(rad)$')
    plt.ylabel(r'$y$')
    plt.legend()  # 添加图例
    plt.annotate("linestyle '" + str(linestyles[i]) + "'", xy=(0.5, -2.5), va='center', ha='left')  # 在图中添加注释


```

## P3 自定义线条宽度
### plt.plot(x, y, linewidth=5) or plt.plot(x, y, lw=5)

```python
N = 50

rows = 2
columns = 2

x = np.linspace(0, 10, N)
y = np.sin(x)**2 + np.cos(x)

grid = plt.GridSpec(rows, columns, wspace=.25, hspace=.25)

linewidth = [2, 3, 4, 5]

plt.figure(figsize=(15, 10))
for i in range(len(linewidth)):
    plt.subplot(grid[i])
    plt.plot(x, y, linestyle = '-.', lw=linewidth[i], label=r'$y = sin^2(x) + cos(x)$')
    plt.axis('equal')
    plt.xlabel(r'$x(rad)$')
    plt.ylabel(r'$y$')
    plt.legend()
    plt.annotate("linewidth = " + str(linewidth[i]), xy=(0.5, -2.5), va='center', ha='left')



```

## P4 创建间隔标记
### 每隔markevery个数据点，做一个标记，P4用圆点做标记

```python
'o'                 # shape for each 5 data
markevery = 10      # mark every
ms = 10             # size of the circle in mark every

N = 50

x = np.linspace(0, 10, N)
y = np.sin(x)**2 + np.cos(x)

plt.figure(figsize=(10,  6))
plt.plot(x, y, 'o', linestyle='-.', linewidth=2, ms=ms, markevery=markevery, label=r'$y=sin^2(x) + cos(x)$')
plt.axis('equal')
plt.xlabel('$x$(rad)')
plt.ylabel('$y$')
plt.legend()
plt.annotate("markevery: " + str(markevery), xy=(0.5, -2.5), va='center', ha='left')







```
