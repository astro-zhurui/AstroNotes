```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True  #设置LaTeX字体为默认字体
plt.style.use('seaborn-white')
```

## P1 多个坐标轴
### ax2 = plt.axes(\[bottom, left, width, height\])
+  bottom：新坐标原点的横坐标 = 默认坐标长度 * bottom 
+ left：新坐标原点的纵坐标 = 默认坐标长度 * left
+ width：横坐标长度 = 原坐标的width倍
+ height：纵坐标长度 = 原坐标的height倍

```python
ax1 = plt.axes()  # 默认坐标轴
ax2 = plt.axes([0.4, 0.6, 0.5, 0.2])  # 第二坐标轴
```

## P2 竖直排列的坐标轴
### 

```python
fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4], ylim=(-1.2, 1.2), xticklabels=[])
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4], ylim=(-1.2, 1.2))

x = np.linspace(0,10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x))
```
