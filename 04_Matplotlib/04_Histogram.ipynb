```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True  #设置LaTeX字体为默认字体
```

## P1 基础用法与细节调整
### plt.hist(x, bins = 10， range=(12, 18), color='darkorange', edgecolor='k')

```python
#------生成N个正态分布的随机数
N = 10000
sigma = 2
mu = 15

np.random.seed(1)
x = np.random.randn(N) * sigma + mu
print(x)

#------绘制直方图
plt.figure(figsize=(16,5))

plt.subplot(1, 2, 1)
plt.hist(x, bins = 10, label=r'$ \mu={}, \sigma={}$'.format(mu, sigma))
#bins表示区间个数
plt.xlabel(r'$x$')
plt.ylabel('number')
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(x, bins = 10, range=(12, 18), color='red', edgecolor='k', label=r'$ \mu={}, \sigma={}$'.format(mu, sigma))
#range选择显示的区间范围，color为直方的颜色，edgecolor='k'显示直方图的边框。
plt.xlabel(r'$x$')
plt.ylabel('number')
plt.legend()
```

## P2 水平直方图

```python
#------生成N个正态分布的随机数
N = 10000
sigma = 2
mu = 15

np.random.seed(1)
x = np.random.randn(N) * sigma + mu
print(x)

#------绘制直方图
plt.hist(x, bins = 25, label=r'$ \mu={}, \sigma={}$'.format(mu, sigma), orientation='horizontal')
#bins表示区间个数
plt.xlabel('number')
plt.ylabel(r'$x$')

plt.legend()
```
